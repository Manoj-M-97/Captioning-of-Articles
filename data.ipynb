```

import _pickle as pickle
import jsonlines
import os
from google.colab import files

heads = []
descs = []
keywords = []

with jsonlines.open('signalmedia-1m.jsonl','r') as reader:
    i = 0
    for obj in reader:
        if i < 1000:
            i += 1
            head = obj["title"]
            desc = [s.strip() for s in obj["content"].splitlines()]
            heads.append(head)
            descs.append(desc)
            keywords.append(None)

        else:
            break
      
# print(heads, descs, keywords)
print(os.path.getsize("tokens.pkl"))
with open('tokens.pkl', 'wb') as f:
    pickle.dump((heads,descs,keywords),f)

json_import.SetContentFile('tokens.pkl')
json_import.Upload()
    
print(os.path.getsize("tokens.pkl"))
print('Uploaded file with ID {}'.format(json_import.get('id')))
```
